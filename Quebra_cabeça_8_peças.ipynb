```python
import random

def move(tab_original):
    """
    Checa possíveis movimentos (os nós)
    da árvore
    """
    movimentos = []
    tab = eval(tab_original)
    i = 0
    j = 0
    while 0 not in tab[i]: i += 1
    j = tab[i].index(0)
    

    if i<2:         #mover o 0 para baixo 
        tab[i][j], tab[i+1][j] = tab[i+1][j], tab[i][j] 
        movimentos.append(str(tab))
        tab[i][j], tab[i+1][j] = tab[i+1][j], tab[i][j]

    if i>0:         #mover o 0 para cima
        tab[i][j], tab[i-1][j] = tab[i-1][j], tab[i][j]  
        movimentos.append(str(tab))
        tab[i][j], tab[i-1][j] = tab[i-1][j], tab[i][j]  

    if j<2:         #mover o 0 para a direita
        tab[i][j], tab[i][j+1] = tab[i][j+1], tab[i][j] 
        movimentos.append(str(tab))
        tab[i][j], tab[i][j+1] = tab[i][j+1], tab[i][j]
    
    if j>0:         #mover o 0 para a esquerda
        tab[i][j], tab[i][j-1] = tab[i][j-1], tab[i][j] 
        movimentos.append(str(tab))
        tab[i][j], tab[i][j-1] = tab[i][j-1], tab[i][j]

    return movimentos

def bfs(start,end):
  
    explorado = []
    banco = [[start]]
    while banco:
        i = 0
        caminho = banco[i]
        banco = banco[:i] + banco[i+1:]
        final = caminho[-1]
        if final in explorado: 
            continue
        for movimento in move(final):
            if movimento in explorado:
                continue
            banco.append(caminho + [movimento])
        explorado.append(final)
        if final == end: break
    return caminho

def bp(start,end):

    explorado = []
    banco = [[start]]
    while banco:
        i = 0
        caminho = banco[i]
        banco = banco[:i+1] + banco[i:]
        final = caminho[-1]
        if final in explorado: 
            continue
        for movimento in move(final):
            if movimento in explorado:
                continue
            banco.append(caminho + [movimento])
        explorado.append(final)
        if final == end: break
    return caminho    

def h_misplaced(tabuleiro):

    #numero_de_movimentos = 0
    misplaced = 0
    comparador = 1
    tab = eval(tabuleiro)
    for i in range(0,3):
        for j in range(0,3):
            if tab[i][j] != comparador:
                misplaced += 1
            comparador += 1
            #numero_de_movimentos += 1
    #print('here')        
    #print(misplaced)
    return misplaced

def gulosa(start,end):

    explorado = []
    banco = [[h_misplaced(start),start]]
    while banco:
        i = 0
        for j in range(1,len(banco)):
            if (banco[i][0]) > (banco[j][0]):
               i = j
        caminho = banco[i]
        banco = banco[:i] + banco[i+1:]
        final = caminho[-1]
        if final in explorado: continue
        for movimento in move(final):
            if movimento in explorado: continue
            novo = [caminho[0] + h_misplaced(movimento) + h_misplaced(final)] + caminho[1:] + [movimento] 
            banco.append(novo)
        explorado.append(final)
        if final == end: break
    return caminho




#tabuleiro2 = str([
#                [4,3,6],
#                [8,7,1],
#                [0,5,2]
#            ])
#print(tabuleiro2)

obj_final = str([
                [1,2,3],
                [4,5,6],
                [7,8,0]
            ])




for rodada in range(99):
  n = 3  
  k = 3   
  m = 0  
  matriz = [] 
  valores = [0,1,2,3,4,5,6,7,8]
  random.shuffle(valores)
  num = valores
  #print(valores)
  for j in range(n):
      linha = []  
      for i in range(m,k):
          #print(num)
          #print(i)
          linha.append(num[i])

      m = k     
      k = k+3
    
      matriz.append(linha)

  print("rodada ", rodada,"\n")    

  tabuleiro = str (matriz)
  print(tabuleiro)

  
  numero_de_movimentos2 = 0
  print("Usando busca gulosa:")
  for i in gulosa(tabuleiro,obj_final):
    numero_de_movimentos2 += 1
    print(i, end="\n")
  print("o numero de movimentos foi ",numero_de_movimentos2)

  print("Usando Busca em Largura:")
  numero_de_movimentos = 0
  for i in bfs(tabuleiro,obj_final):
    numero_de_movimentos += 1
    print(i, end="\n")
  print("o numero de movimentos foi ",numero_de_movimentos)

  #numero_de_movimentos3 = 0
  #print("Usando busca em profundidade:")
  #for i in bp(tabuleiro,obj_final):
  #  numero_de_movimentos3 += 1
  #  print(i, end="\n")
  #print("o numero de movimentos foi ",numero_de_movimentos3)
```

Mediante aos experimentos pode se observar algumas coisas em relação a otimalidade, tempo, memoria e completeza.

Em relação a *memoria* os algoritmos de busca em largura e profundidade consomem mais memória, mas especificamente nessa ordem, busca em profundidade, busca em largura e guloso, que por sua vez é o que consome menos memória.

Ja em relação ao *tempo*, os algoritmos estão ranqueados na mesma ordem, a busca em profundidade por sua vez foi o unico que algumas vezes não conseguiu encontrar a solução e o guloso mesmo que com uma quantidade maior de movimentos sempre acha a solução de maneira mais rápida.

Quanto a *otimalidade*, o algoritmo de busca em largura é o que apresenta uma solução ótima, levando em conta que sempre possui o caminho menos custoso entre os outros algoritmos.

E por fim em relação a *completeza* os algoritmos guloso e o  de busca em largura são completos, pois sempre que uma solução existe temos a garantia de encontra-la, ja o de busca em profundidade nao temos tal garantia, sendo assim incompleto.
