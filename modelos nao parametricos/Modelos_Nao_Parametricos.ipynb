```python
!pip install -U -q PyDrive
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt
from sklearn.preprocessing import Normalizer

import seaborn as sns

from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn import metrics
from mlxtend.plotting import plot_decision_regions

from sklearn.model_selection import cross_val_score

#xpoints = np.array([0, 5])
#ypoints = np.array([0, 250])
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
#Classificacao
print('\n')
print('classificacao')
print('\n')
print('')
#https://drive.google.com/file/d/1dTsCI756okOl9kCUVE5xjtrbtuu74U1q/view?usp=sharing
fileDownloaded = drive.CreateFile({'id':'1dTsCI756okOl9kCUVE5xjtrbtuu74U1q'})
fileDownloaded.GetContentFile('geral_vacinaja_classificacao.csv')
df = pd.read_csv('geral_vacinaja_classificacao.csv', delimiter=',' )
#print(df)
#print(df.size)

#normalizacao


cols = ['A','B','C','D','E','F']
nmp=df.to_numpy()
names =['BOM', 'RUIM']
#print(scaler.inverse_transform(scaled_data))
kf = KFold(n_splits=5)
data = pd.DataFrame(nmp,columns=cols)
scaler = MinMaxScaler(feature_range=(0,1))
cols1= ['A','B','C','D','E']
scaled_data = scaler.fit_transform(data[cols1])
scaler.inverse_transform(scaled_data)

X = scaler.inverse_transform(scaled_data)[:,[0,4]]
y = data['F'].to_numpy()
clf = svm.SVC(kernel='poly')
clf.fit(X, y)
#svm.SVC()

#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
#h = .02
for train, test in kf.split(nmp):  
  clf.fit(X, y)
  plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

  # plot the decision function
  ax = plt.gca()
  xlim = ax.get_xlim()
  ylim = ax.get_ylim()

  # create grid to evaluate model
  xx = np.linspace(xlim[0], xlim[1], 30)
  yy = np.linspace(ylim[0], ylim[1], 30)
  YY, XX = np.meshgrid(yy, xx)
  xy = np.vstack([XX.ravel(), YY.ravel()]).T
  Z = clf.decision_function(xy).reshape(XX.shape)

  # plot decision boundary and margins
  ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
            linestyles=['--', '-', '--'])
  # plot support vectors
  ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
            linewidth=1, facecolors='none', edgecolors='k')
  
plt.title("SVM (kernel = poly)")                                                                  
plt.show()

clf = svm.SVC(kernel='rbf')  
for train, test in kf.split(nmp):  
  clf.fit(X, y)
  plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

  # plot the decision function
  ax = plt.gca()
  xlim = ax.get_xlim()
  ylim = ax.get_ylim()

  # create grid to evaluate model
  xx = np.linspace(xlim[0], xlim[1], 30)
  yy = np.linspace(ylim[0], ylim[1], 30)
  YY, XX = np.meshgrid(yy, xx)
  xy = np.vstack([XX.ravel(), YY.ravel()]).T
  Z = clf.decision_function(xy).reshape(XX.shape)

  # plot decision boundary and margins
  ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
            linestyles=['--', '-', '--'])
  # plot support vectors
  ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
            linewidth=1, facecolors='none', edgecolors='k')
  
plt.title("SVM (kernel = rbf)")                                                                  
plt.show()

clf = svm.SVC(kernel='sigmoid')  
for train, test in kf.split(nmp):  
  clf.fit(X, y)
  plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

  # plot the decision function
  ax = plt.gca()
  xlim = ax.get_xlim()
  ylim = ax.get_ylim()

  # create grid to evaluate model
  xx = np.linspace(xlim[0], xlim[1], 30)
  yy = np.linspace(ylim[0], ylim[1], 30)
  YY, XX = np.meshgrid(yy, xx)
  xy = np.vstack([XX.ravel(), YY.ravel()]).T
  Z = clf.decision_function(xy).reshape(XX.shape)

  # plot decision boundary and margins
  ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
            linestyles=['--', '-', '--'])
  # plot support vectors
  ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
            linewidth=1, facecolors='none', edgecolors='k')
  
plt.title("SVM (kernel = sigmoid)")                                                                  
plt.show()

  
```
